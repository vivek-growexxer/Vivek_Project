```python
import pandas as pd
```

```python
columns_max = ['train_accuracy', 'test_accuracy', 'roc_auc',
           'precision_0', 'recall_0', 'f1_0', 
           'precision_1', 'recall_1', 'f1_1',
           'ks_stat','tp', 'tn']

# columns_max = ['train_accuracy', 'test_accuracy', 'roc_auc', 
#                'f1_0', 'f1_1', 'ks_stat','tp', 'tn']

# columns_max = ['tp', 'tn']

columns_min = ['fp', 'fn']
top_n = 30
```

### Using data from Approach 1

```python
app_1 = pd.read_csv("./adasyn_app1_eval.csv")
app_1.head()
```

```python
all_col_top_rows = {}
for col in columns_max:
    top_rows = app_1.sort_values(by=[col],ascending=False)[:top_n]['k_neighbour']
    all_col_top_rows[col + '_top_rows'] = set(top_rows)

for col in columns_min:
    top_rows = app_1.sort_values(by=[col],ascending=True)[:top_n]['k_neighbour']
    all_col_top_rows[col + '_top_rows'] = set(top_rows)
```

```python
# Finding rows which are common in all the top_rows
common_rows = set()
for key,value in all_col_top_rows.items():
    if len(common_rows) == 0:
        common_rows = value
    else:
        common_rows = common_rows.intersection(value)
#     break
```

```python
common_rows
```

```python
top_rows_app_1 = app_1[app_1['k_neighbour'].isin(common_rows)]
# top_rows_app_1
```

```python
top_rows_app_1.sort_values(by=['test_accuracy'],ascending=False)
```
